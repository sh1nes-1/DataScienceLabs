### 5

Зверніть увагу, що стовпець `DebtRatio` містить неправдиві дані. Тільки значення, які відповідають відомому доходу за місяць, є відносними. Решта – абсолютні значення виплат відсотків за місяць. 

Виправте дані, зробивши всі значення стовпця `DebtRatio` абсолютними (помножте їх на `MonthlyIncome`).  

#### *Функції, які можуть бути корисними при рішенні:*

Звернення до елементів DataFrame:
  * елемент: `data.loc[i, 'назваСтовпця']`
  * стовпець: `data['назваСтовпця']`
  * підматриця: `data.loc[a:b, списокНазвСтовпців]`

Умовна індексація:
* `data.loc[data['стовпець'] > 20, списокНазвСтовпців]`

або краще так:

* `i = data['столбец'] > 20`  # вектор `True` і `False`
* `data.loc[i, 'назваСтовпця']`

У підматриць номери рядків спадкуються від первісного.

* `pandas.isnull(скаляр або масив)` - перевірка, чи є значення невизначеним (`NaN`)
* `pandas.notnull(скаляр або масив)` - перевірка, чи є значення визначеним (не `NaN`)

```python
not_null_income_rows = pd.notnull(data['MonthlyIncome'])
data.loc[not_null_income_rows, 'DebtRatio'] *= data.loc[not_null_income_rows, 'MonthlyIncome']
data.head(10)
```

### 7

Визначте середній щомісячний дохід і присвойте всім клієнтам з невідомим доходом отримане значення.

*Функції, що будуть корисними при рішенні:  `.mean()`*

*Інші статистики:* https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats

```python
null_income_rows = pd.isnull(data['MonthlyIncome'])
data.loc[null_income_rows, 'MonthlyIncome'] = data['MonthlyIncome'].mean()
data.head(10)
```

### 9a

Побудуйте графік розсіювання на осях `age` і `Debt`. Cинім відмітьте клієнтів без великої заборгваності (`SeriousDlqin2yrs = 0`) і червоним — боржників (`SeriousDlqin2yrs = 1`).

Візуалізуйте попарні залежності між ознаками `'age', 'MonthlyIncome', 'NumberOfDependents'`. Обмежте при цьому дохід значенням 25000.

Які закономірності можна спостерігати на  графіках?
